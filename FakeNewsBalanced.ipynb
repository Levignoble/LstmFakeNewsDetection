# LSTM


# Building And Training LSTM Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Embedding, LSTM
# Define embedding vector features
embedding_vector_features = 40
# Initialize the model
model = Sequential()
# Add the embedding layer with explicit input_length
model.add(Embedding(input_dim=total_words,
                    output_dim=embedding_vector_features,
                    input_length=42))  # Explicitly define input_length
# Add dropout to prevent overfitting
model.add(Dropout(0.3))
# Add LSTM layer
model.add(LSTM(100))
# Add another dropout layer
model.add(Dropout(0.3))
# Add final dense layer with sigmoid activation for binary classification
model.add(Dense(1, activation='sigmoid'))
model.build(input_shape=(None, 42))  # Batch size is None, sequence length is 42
# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# Print the model summary
model.summary()
```

```python
y_train = np.asarray(train_df['true'])
y_val = np.asarray(val_df['true'])

# Training the model
model.fit(padded_train, y_train, batch_size = 64, validation_data=(padded_val, y_val), epochs = 3)
model.export('/content/drive/My Drive/lstm.keras')
```

# Assessing Trained Model Performance

```python
# Making predictions
predictions = model.predict(padded_test)

# Convert probabilities to binary class labels (0 or 1)
prediction_classes = (predictions > 0.5).astype("int32")

# Getting the accuracy
from sklearn.metrics import accuracy_score
y_test = np.asarray(test_df['true'])

accuracy = accuracy_score(y_test, prediction_classes)
print("LSTM Model Accuracy:", accuracy)

# Getting the confusion matrix
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_test, prediction_classes)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

# Getting the classification report
from sklearn.metrics import classification_report
print(classification_report(y_test, prediction_classes))
```

# Bidirectional LSTM

# Building And Training Bidirectional LSTM Model

```python
from keras.models import Sequential
from keras.layers import Embedding, Dropout, Bidirectional, LSTM, Dense, Input

# Define embedding vector features
embedding_vector_features = 40
input_length = 20  # Matches the maximum padded sequence length

# Initialize the model
model = Sequential()

# Add input layer for clear shape inference
model.add(Input(shape=(input_length,)))

# Add embedding layer
model.add(Embedding(input_dim=total_words,  # Total vocabulary size
                    output_dim=embedding_vector_features,
                    input_length=input_length))

# Add dropout layer
model.add(Dropout(0.3))

# Add bidirectional LSTM layer
model.add(Bidirectional(LSTM(100)))

# Add another dropout layer
model.add(Dropout(0.3))

# Add dense layer for binary classification
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print the model summary
print(model.summary())

```

```python
y_train = np.asarray(train_df['true'])
y_val = np.asarray(val_df['true'])

# Training the model
model.fit(padded_train, y_train, batch_size = 64, validation_data=(padded_val, y_val), epochs = 3)
model.save('/content/drive/My Drive/bilstm.keras')
```

# Assessing Trained Model Performance

```python
# Making predictions
predictions = model.predict(padded_test)

# Convert probabilities to binary class labels (0 or 1)
prediction_classes = (predictions > 0.5).astype("int32")

# Getting the accuracy
y_test = np.asarray(test_df['true'])

accuracy = accuracy_score(y_test, prediction_classes)
print("Bidirectional LSTM Model Accuracy:", accuracy)

# Getting the confusion matrix
cm = confusion_matrix(y_test, prediction_classes)
plt.figure(figsize=(6, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")

# Getting the classification report
print(classification_report(y_test, prediction_classes))
```

# CNN-BiLSTM

# Building And Training CNN Model

```python
## Creating model Using CNN
from keras.layers import Conv1D, MaxPool1D

embedding_vector_features=40
model=Sequential()
model.add(Embedding(total_words,embedding_vector_features,input_length=20))

model.add(Dropout(0.3))

model.add(Conv1D(32, 5, activation='relu'))
model.add(MaxPool1D())

model.add(Conv1D(32, 5, activation='relu'))
model.add(MaxPool1D())

model.add(Bidirectional(LSTM(100)))
model.add(Dropout(0.3))

model.add(Dense(1,activation='sigmoid'))
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
print(model.summary())
```
